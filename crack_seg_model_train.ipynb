```python
from ultralytics import YOLO

# YOLOv8m 모델 로드
model = YOLO('yolov8m-seg.pt')  # 세그멘테이션을 위한 pre-trained 모델 사용

# 데이터셋 경로 설정
data_yaml = '/content/drive/MyDrive/crack.v2i.yolov8/data.yaml'

# 모델 훈련
model.train(data=data_yaml, epochs=100, imgsz=640, batch=16, name='crack_detection_yolov8m')

# 모델 검증
model.val(data=data_yaml)

# 훈련된 모델 저장
model.save('/content/drive/MyDrive/crack_detection_yolov8m.pt')
```
